## Interpolation Problem Statement

The Interpolation Problem Statement involves finding a function ŷ(x) that precisely passes through a given set of data points (xᵢ,yᵢ), where i = 1,2,...,n. This function ensures that for every point xᵢ in the dataset, ŷ(xᵢ)= yᵢ, allowing it to predict values between known data points. Unlike regression analysis, which relies on a presupposed model, interpolation focuses on the inherent characteristics of the data itself, making it suitable for scenarios requiring smooth transitions and exact matches at given points, such as in scientific data analysis and computer graphics processing.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

# Example data points
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 0.8, 0.9, 0.1, -0.8, -1])

# Creating linear and cubic interpolation functions
linear_interp = interp1d(x, y)
cubic_interp = interp1d(x, y, kind='cubic')

# New x values for interpolation
x_new = np.linspace(0, 5, 100)

# Interpolated y values
y_linear = linear_interp(x_new)
y_cubic = cubic_interp(x_new)

# Plotting
plt.figure(figsize=(12, 6))

# Original data points
plt.scatter(x, y, label='Data Points')

# Linear Interpolation
plt.plot(x_new, y_linear, label='Linear Interpolation')

# Cubic Interpolation
plt.plot(x_new, y_cubic, label='Cubic Interpolation')

plt.title('Interpolation Example')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

## Linear Interpolation

The estimated point in linear interpolation is thought to reside on the line connecting the closest points to the left and right.
Then the linear interpolation at x is: ŷ(x) = yᵢ + (y_i+1 - yᵢ)(x -xᵢ)/(x_i+1 - xᵢ)

```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')

# Updated data points
x_new = [0, 2, 4]
y_new = [2, 3, 5]

# Create a new linear interpolation function
f_new = interp1d(x_new, y_new)

# Compute the interpolated value at x = 3
y_hat_new = f_new(3)
print("Interpolated value at x = 3:", y_hat_new)

# Plotting with the updated data points and the interpolated point
plt.figure(figsize=(10, 8))
plt.plot(x_new, y_new, '-og', label='Updated Data')  # Using green color for new data
plt.plot(3, y_hat_new, 'r*', label='Interpolated at x=3')  # Using a red star for the interpolated point
plt.title('Linear Interpolation at x = 3')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

## Cubic Spline Interpolation

A collection of piecewise cubic functions serves as the interpolating function in cubic spline interpolation, as seen in the accompanying picture.

```python
from scipy.interpolate import CubicSpline
import numpy as np
import matplotlib.pyplot as plt

# Updated data points
x_updated = np.array([0, 2, 4, 6])
y_updated = np.array([3, 2, 0, 1])

# Create a cubic spline interpolation function with 'natural' boundary conditions
f_cubic = CubicSpline(x_updated, y_updated, bc_type='natural')

# Generating new x values for interpolation
x_new_cubic = np.linspace(0, 6, 100)
y_new_cubic = f_cubic(x_new_cubic)

# Plotting
plt.figure(figsize=(10, 8))
plt.plot(x_new_cubic, y_new_cubic, 'b', label='Cubic Spline Interpolation')
plt.plot(x_updated, y_updated, 'ro', label='Data Points')
plt.title('Cubic Spline Interpolation with Updated Data')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
