#### Continuity
Continuity in mathematics refers to the property of a function where its graph can be drawn without lifting the pen from the paper. In more technical terms, a function $f(x)$ is continuous at a point $x=a$ if the following three conditions are met:

Function is Defined: The function $f(a)$ is defined at $a$.

Limit Exists: The limit of $f(x)$ as $x$ approaches $a$ exists.
Limit Equals Function Value: The limit of $f(x)$ as $x$ approaches $a$ is equal to $f(a)$.

Mathematically, this is expressed as: $lim_{x→a}f(x)=f(a)$

#### Differentiation
Differentiation, on the other hand, is a process in mathematics that deals with the rate at which a function is changing at any given point. The derivative of a function at a point gives the slope of the tangent to the function's graph at that point. If $f(x)$ is a function, then its derivative $f'(x)$ is defined as the limit:
$f'(x) = lim_{h→0}\frac{f(x+h)-f(x)}{h}$. A function is differentiable at a point $x=a$ if this limit exists.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x) = x^2 and its derivative
def f(x):
    return x ** 2

def derivative_f(x):
    return 2 * x

# Choose a point (for example, x = 2)
x_point = 2
y_point = f(x_point)
slope_at_point = derivative_f(x_point)

# Creating points for the function plot
x_values = np.linspace(-3, 3, 400)
y_values = f(x_values)

# Creating points for the tangent line at x = 2
# Tangent line equation: y = slope * (x - x_point) + y_point
tangent_line = slope_at_point * (x_values - x_point) + y_point

# Plotting the function and the tangent line
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='f(x) = x^2', color='blue')
plt.plot(x_values, tangent_line, label='Tangent at x = 2', color='red', linestyle='--')
plt.scatter([x_point], [y_point], color='green')  # Point of tangency

# Adding labels and title
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Function f(x) = x^2 and its Tangent Line at x = 2')
plt.legend()
plt.grid(True)
plt.show()
```

Rolle theorm: Let f : [a,b] → R be a continuous function and assume that its derivative exists on (a,b). If f(a) = f(b), then there is a < c < b such that f'(c) = 0.

Mean value theorm: Let f:[a,b]→ R be a continuous function and assume that its derivative exists on (a,b). Then there is a < c < b such that f(b)=f(a)+(b-a)f'(c)

***The second derivative*** in mathematics refers to the derivative of the derivative of a function. Essentially, if $f(x)$ is a function, its first derivative $f'(x)$ describes the instantaneous rate of change of $f(x)$ (i.e., the slope), and the second derivative $f"(x)$ describes how this rate of change is changing.

Key aspects of the second derivative include:

Concavity: The second derivative tells us about the concavity of the function. If $f"(x) > 0$, the function is concave up (curves upwards) in that region; if $f"(x) < 0$, it is concave down (curves downwards).

Inflection Points: These are points where the concavity of the function changes, i.e., where the second derivative changes sign from positive to negative or vice versa. At these points, the curve of the function changes from curving upwards to downwards, or the opposite.

```python
# Define the second derivative of f(x) = x^2
def second_derivative_f(x):
    return 2

# Creating points for the first derivative plot
first_derivative_values = derivative_f(x_values)

# Creating points for the second derivative plot (constant value)
second_derivative_values = np.full_like(x_values, second_derivative_f(x_values))

# Plotting the function, first derivative, and second derivative
plt.figure(figsize=(12, 8))

# Plot for the original function f(x) = x^2
plt.subplot(3, 1, 1)
plt.plot(x_values, y_values, label='f(x) = x^2', color='blue')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Function f(x) = x^2')
plt.grid(True)

# Plot for the first derivative f'(x) = 2x
plt.subplot(3, 1, 2)
plt.plot(x_values, first_derivative_values, label="f'(x) = 2x", color='green')
plt.xlabel('x')
plt.ylabel("f'(x)")
plt.title("First Derivative f'(x) = 2x")
plt.grid(True)

# Plot for the second derivative f''(x) = 2
plt.subplot(3, 1, 3)
plt.plot(x_values, second_derivative_values, label="f''(x) = 2", color='red')
plt.xlabel('x')
plt.ylabel("f''(x)")
plt.title("Second Derivative f''(x) = 2")
plt.grid(True)

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

***Taylor's Theorem ***is a fundamental concept in mathematics, providing a method for approximating smooth functions with polynomials. Essentially, it states that any sufficiently smooth function can be approximated near a certain point by a polynomial.

Taylor Series
At the heart of Taylor's Theorem is the Taylor series. If a function $f(x)$ is differentiable near a point $a$, then it can be represented as:$f(x)  \approx f(a) +f'(a)(x-a) + \frac{f"(a)}{2!}(x-a)^2 + \frac{f'''(a)}{3!}(x-a)^3 + ...$ Here, $f'(a),f''(a),f'''(a)$,… are the first, second, third, etc., derivatives of the function at point $a$, and $n!$ denotes the factorial of $n$.

Advantages of Using Taylor Series
Function Approximation: Taylor series can be used to approximate complex functions, which is extremely useful in engineering and physics, especially when an analytical solution is not available or is too complex.
Understanding Function Behavior: Taylor series provide a way to understand the overall properties of a function by observing its behavior around a certain point.

```python
import math

# Define the function e^x and its Taylor series approximation up to the third order
def e_x(x):
    return np.exp(x)

def taylor_e_x(x, n_terms=4):
    # Taylor series expansion at a = 0
    taylor_series = sum([(x**i) / math.factorial(i) for i in range(n_terms)])
    return taylor_series

# Creating points for the plot
x_values_taylor = np.linspace(-2, 2, 400)
y_values_actual = e_x(x_values_taylor)
y_values_taylor = taylor_e_x(x_values_taylor)

# Plotting the actual function and its Taylor approximation
plt.figure(figsize=(8, 6))
plt.plot(x_values_taylor, y_values_actual, label='Actual e^x', color='blue')
plt.plot(x_values_taylor, y_values_taylor, label='Taylor Series (up to 3rd order)', color='red', linestyle='--')
plt.scatter([0], [1], color='green')  # Mark the point of expansion

# Adding labels and title
plt.xlabel('x')
plt.ylabel('Value')
plt.title('Function e^x and its Taylor Series Approximation at x = 0')
plt.legend()
plt.grid(True)
plt.show()
```

#### Optimization with gradient descent

Gradient Descent is a fundamental optimization technique especially prevalent in machine learning and deep learning. It's utilized to minimize a function, essentially finding the parameters which result in the function's lowest value.

Key Concepts in Gradient Descent Optimization

Global Minimizer:

Definition: For a function $f:R^d → R$, a point $x^* ∈ R^d$ is a global minimizer if $f(x) >= f(x^*)$ for all $x ∈ R^d$. A global minimizer is the point where the function reaches its absolute lowest value.

Local Minimizer:

Definition: For $f:R^d → R$, a point $x^* ∈ R^d$ is a local minimizer if there exists $δ > 0$ such that $f(x) >= f(x^*)$ for all $x$ in the neighborhood $ B_δ(x^*) \backslash {x^*} $. A local minimizer refers to a point where the function reaches a minimum, compared to its immediate surroundings.

Descent Direction:

Definition: For $f:R^d → R$, a vector $v$ is a descent direction at $x_0$ if there exists $α^* > 0$ such that $f(x_0 + αv) < f(x_0)$ for all $α ∈ (0,α^*)$. Identifying a descent direction helps in determining which way to move to decrease the function's value.

First-Order Necessary Condition:

Definition: If $f:R^d → R$ is continuously differentiable, and $x_0$ is a local minimizer, then $∇f(x_0) = 0$. This condition helps in identifying potential local minimizers.

Second-Order Necessary Condition:

Definition: If $f:R^d → R$ is twice continuously differentiable, and $x_0$ is a local minimizer, then the Hessian matrix $H_f(x_0)$ is positive semi-definite (PSD). This condition provides a way to test whether a stationary point (where the first derivative is zero) is a local minimizer.

Convex Set:

Definition: A set $D ⊆ R^d$ is convex if for all $x,y ∈ D$ and all $α ∈[0,1]$, the point $(1-α)x+αy$ is also in $D$. Convex sets are fundamental in optimization as they ensure no local minima other than the global minimum.

Convex Function:

Definition: A function $f:R^d → R$ is convex if, for all $x,y∈R$ and all $α∈[0,1]$,$f((1-α)x+αy)≤(1-α)f(x)+αf(y)$. Convex functions simplify optimization because they ensure any local minimum is a global minimum.

Gradient Descent Steps

Initialize Parameters: Start with random values.

Compute Gradient: Find the slope of the objective function at current parameters.

Update Parameters: Adjust parameters in the opposite direction of the gradient, scaled by the learning rate.

Iterate: Repeat until a stopping criterion, like a sufficiently small gradient or a maximum number of steps, is met.

```python
def f(x):
    """ The function to be minimized, f(x) = x^2 """
    return x ** 2

def gradient_f(x):
    """ The gradient of the function, f'(x) = 2x """
    return 2 * x

def gradient_descent(starting_x, learning_rate, n_steps):
    """ Gradient Descent Algorithm """
    x = starting_x
    x_history = [x]
    f_history = [f(x)]

    for _ in range(n_steps):
        grad = gradient_f(x)
        x = x - learning_rate * grad  # Update rule
        x_history.append(x)
        f_history.append(f(x))

    return x_history, f_history

# Parameters for Gradient Descent
starting_x = 10  # Starting point
learning_rate = 0.1  # Learning rate
n_steps = 20  # Number of steps

# Running Gradient Descent
x_history, f_history = gradient_descent(starting_x, learning_rate, n_steps)

# Plotting the function and the updates
x_values = np.linspace(-11, 11, 400)
y_values = f(x_values)

plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='f(x) = x^2', color='blue')
plt.scatter(x_history, f_history, color='red', marker='o')
plt.title('Gradient Descent on f(x) = x^2')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
